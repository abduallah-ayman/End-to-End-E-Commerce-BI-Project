```python
import pandas as pd
import random
from faker import Faker
```

```python
fake = Faker()

NUM_INVENTORY = 5024
```

```python
movements = []

for inventory_id in range(1, NUM_INVENTORY + 1):

    num_movements = random.randint(5, 30)

    for _ in range(num_movements):
        movement_type = random.choice(["IN", "OUT"])

        if movement_type == "IN":
            qty_change = random.randint(1, 50)
        else:
            qty_change = -random.randint(1, 30)

        movement_date = fake.date_time_between(
            start_date="-1y",
            end_date="now"
        )

        movements.append({
            "InventoryID": inventory_id,
            "QuantityChange": qty_change,
            "MovementType": movement_type,
            "MovementDate": movement_date
        })

df = pd.DataFrame(movements)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Inventory\InventoryMovement.xlsx"
df.to_excel(output_path, index=False)
```
