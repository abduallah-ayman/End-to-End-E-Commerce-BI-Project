```python
from faker import Faker
import pandas as pd
```

```python
fake = Faker()

NUM_SELLERS = 200
addresses = []
```

```python
for seller_id in range(1, NUM_SELLERS + 1):
    addresses.append({
        "SellerID": seller_id,
        "Street": fake.street_address(),
        "City": fake.city(),
        "Country": fake.country()
    })

df = pd.DataFrame(addresses)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Seller\SellerAddress.xlsx"
df.to_excel(output_path, index=False)
```
