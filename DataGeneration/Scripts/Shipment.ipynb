```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
NUM_ORDERS = 800_000
NUM_SHIPMENT_TYPES = 4

DELIVERY_PARTNERS = ["DHL", "FedEx", "Aramex", "UPS", "Local Courier"]

MAX_DAYS_BY_TYPE = {
    1: 7,    # Standard
    2: 3,    # Express
    3: 0,    # Same Day
    4: 14    # Economy
}

START_ORDER_DATE = datetime(2023, 1, 1)
```

```python
shipments = []

for order_id in range(1, NUM_ORDERS + 1):
    # Simulated order date (you may remove if already exists)
    order_date = START_ORDER_DATE + timedelta(days=random.randint(0, 730))

    shipment_type = random.randint(1, NUM_SHIPMENT_TYPES)
    shipment_date = order_date + timedelta(days=random.randint(0, 2))

    delivery_date = shipment_date + timedelta(
        days=random.randint(0, MAX_DAYS_BY_TYPE[shipment_type] + 2)
    )

    shipments.append({
        "OrderID": order_id,
        "ShipmentTypeID": shipment_type,
        "DeliveryPartner": random.choice(DELIVERY_PARTNERS),
        "ShipmentDate": shipment_date.date(),
        "DeliveryDate": delivery_date.date(),
        "Status": "Delivered",
        "CustomerRating": random.randint(1, 5)
    })

df_shipments = pd.DataFrame(shipments)
```
