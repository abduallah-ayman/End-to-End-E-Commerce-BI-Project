```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()

NUM_CUSTOMERS = 50000
addresses = []
```

```python
for customer_id in range(1, NUM_CUSTOMERS + 1):  # 1 to 50000
    # Always a Billing address
    addresses.append({
        "CustomerID": customer_id,
        "AddressType": "Billing",
        "Street": fake.street_address(),
        "City": fake.city(),
        "State": fake.state(),
        "PostalCode": fake.postcode(),
        "Country": fake.country()
    })

    # 40% chance for Shipping address
    if random.random() < 0.4:
        addresses.append({
            "CustomerID": customer_id,
            "AddressType": "Shipping",
            "Street": fake.street_address(),
            "City": fake.city(),
            "State": fake.state(),
            "PostalCode": fake.postcode(),
            "Country": fake.country()
        })

df = pd.DataFrame(addresses)
```

```python
# Save to Excel
output_path = "D:\Portfolio\Data generation\Data\Customer\CustomerAddress.xlsx"
df.to_excel(output_path, index=False)
```
