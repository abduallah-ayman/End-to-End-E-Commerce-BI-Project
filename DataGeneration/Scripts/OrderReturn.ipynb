```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta
```

```python
fake = Faker()

NUM_RETURNS = 100000  # Adjust number of returns (example: 100k)
MAX_ORDERITEM_ID = 1500000 

reasons = [
    "Damaged product",
    "Wrong product",
    "Late delivery",
    "Changed mind",
    "Other"
]
```

```python
returns = []

for _ in range(NUM_RETURNS):
    order_item_id = random.randint(1, MAX_ORDERITEM_ID)
    return_date = fake.date_time_between(
        start_date="-1y",
        end_date="now"
    ).date()

    returns.append({
        "OrderItemID": order_item_id,
        "Reason": random.choice(reasons),
        "ReturnDate": return_date
    })

df_returns = pd.DataFrame(returns)
```
