```python
from faker import Faker
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
fake = Faker()

NUM_SELLERS = 200
contracts = []
```

```python
for seller_id in range(1, NUM_SELLERS + 1):
    start_date = fake.date_between(start_date='-5y', end_date='today')
    
    # 50% chance of NULL (ongoing contract)
    if random.random() < 0.5:
        end_date = None
    else:
        # End date between start_date + 1 month and today + 1 year
        end_date = start_date + timedelta(days=random.randint(30, 365))
    
    commission_rate = round(random.uniform(5.0, 20.0), 2)
    
    contracts.append({
        "SellerID": seller_id,
        "StartDate": start_date,
        "EndDate": end_date,
        "CommissionRate": commission_rate
    })

df = pd.DataFrame(contracts)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Seller\SellerContract.xlsx"
df.to_excel(output_path, index=False)
```
