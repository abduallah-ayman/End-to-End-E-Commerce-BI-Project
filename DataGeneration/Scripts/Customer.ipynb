```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
```

```python
NUM_CUSTOMERS = 50000

segment_weights = {
    1: 0.55,  # Regular
    2: 0.20,  # Silver
    3: 0.15,  # Gold
    4: 0.08,  # Platinum
    5: 0.02   # Corporate
}
segment_ids = list(segment_weights.keys())
weights = list(segment_weights.values())

customers = []
```

```python
for _ in range(NUM_CUSTOMERS):
    gender = random.choice(["M", "F"])
    first_name = fake.first_name_male() if gender == "M" else fake.first_name_female()
    last_name = fake.last_name()
    
    customers.append({
        "FirstName": first_name,
        "LastName": last_name,
        "Gender": gender,
        "DateOfBirth": fake.date_of_birth(minimum_age=18, maximum_age=70),
        "SegmentID": random.choices(segment_ids, weights=weights, k=1)[0]
    })

df = pd.DataFrame(customers)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Customer\Customer.xlsx"
df.to_excel(output_path, index=False)
```
