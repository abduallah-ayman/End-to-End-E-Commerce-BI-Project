```python
import pandas as pd
import random
```

```python
NUM_ORDERS = 800000
```

```python
discounts = []

for order_id in range(1, NUM_ORDERS + 1):
    # Randomly decide if this order has a discount
    has_discount = random.choices([True, False], weights=[20, 80])[0]  # 30% chance
    discount_amount = round(random.uniform(5, 100), 2) if has_discount else 0

    discounts.append({
        "OrderID": order_id,
        "DiscountAmount": discount_amount
    })

df_discounts = pd.DataFrame(discounts)
```

```python
# Save to Excel
output_path = "D:\Portfolio\Data generation\Data\Order\OrderDiscount.xlsx"
df_discounts.to_excel(output_path, index=False)
```
