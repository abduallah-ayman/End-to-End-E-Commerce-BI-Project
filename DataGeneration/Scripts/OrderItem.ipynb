```python
import pandas as pd
import random
import numpy as np
```

```python
NUM_ORDERS = 800000
NUM_PRODUCTS = 2494
```

```python
order_items = []

for order_id in range(1, NUM_ORDERS + 1):
    num_items = random.randint(1, 6)

    # Avoid duplicate products per order
    product_ids = random.sample(
        range(1, NUM_PRODUCTS + 1),
        k=num_items
    )

    for product_id in product_ids:
        order_items.append({
            "OrderID": order_id,
            "ProductID": product_id,
            "Quantity": random.randint(1, 5),
            "UnitPrice": 0
        })

df = pd.DataFrame(order_items)
```

```python
df
```

```python
df1, df2, df3 = np.array_split(df, 3)

print(len(df1), len(df2), len(df3))
```
