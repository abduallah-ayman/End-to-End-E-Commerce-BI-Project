```python
from faker import Faker
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
fake = Faker()

NUM_CUSTOMERS = 50000
login_records = []
```

```python
for customer_id in range(1, NUM_CUSTOMERS + 1):
    num_logins = random.randint(5, 20)
    for _ in range(num_logins):
        # Random date in the last 3 years
        start_date = datetime.now() - timedelta(days=3*365)
        login_time = fake.date_time_between(start_date=start_date, end_date='now')
        
        login_records.append({
            "CustomerID": customer_id,
            "LoginTime": login_time,
            "IPAddress": fake.ipv4()
        })

df = pd.DataFrame(login_records)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Customer\CustomerLoginHistory.xlsx"
df.to_excel(output_path, index=False)
```
