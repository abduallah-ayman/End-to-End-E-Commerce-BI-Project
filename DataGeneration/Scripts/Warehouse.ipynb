```python
from faker import Faker
import pandas as pd
```

```python
fake = Faker()

NUM_WAREHOUSES = 20
```

```python
warehouses = []

for i in range(1, NUM_WAREHOUSES + 1):
    warehouses.append({
        "WarehouseName": f"Warehouse {i}",
        "City": fake.city(),
        "Country": fake.country()
    })

df = pd.DataFrame(warehouses)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Inventory\Warehouse.xlsx"
df.to_excel(output_path, index=False)
```
