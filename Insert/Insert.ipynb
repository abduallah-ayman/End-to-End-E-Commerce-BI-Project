```python
import pyodbc
import pandas as pd
```

```python
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-S0HVABM\MSSQLSERVER22;"
    "DATABASE=E_Commerce;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()
cursor.fast_executemany = True
```

```python
def insert_dataframe(df, table_name, cursor, conn, batch_size=5000):
    columns = ",".join(df.columns)
    placeholders = ",".join("?" * len(df.columns))
    
    sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"

    data = df.values.tolist()

    for i in range(0, len(data), batch_size):
        batch = data[i:i+batch_size]
        cursor.executemany(sql, batch)
        conn.commit()

    print(f"Inserted {len(df)} rows into {table_name}")
```

```python
df = pd.read_excel("D:\Portfolio\Data generation\Data\Customer\CustomerSegment.xlsx")

insert_dataframe(
    df=df,
    table_name="core.CustomerSegment",
    cursor=cursor,
    conn=conn
)
```
