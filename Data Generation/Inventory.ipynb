```python
import pandas as pd
import random
```

```python
NUM_PRODUCTS = 2494
NUM_WAREHOUSES = 20
```

```python
inventory_records = []

for product_id in range(1, NUM_PRODUCTS + 1):
    # Each product stored in 1â€“3 warehouses
    num_warehouses = random.randint(1, 3)
    warehouses = random.sample(range(1, NUM_WAREHOUSES + 1), num_warehouses)

    for warehouse_id in warehouses:
        inventory_records.append({
            "ProductID": product_id,
            "WarehouseID": warehouse_id,
            "QuantityAvailable": random.randint(0, 1000)
        })

df = pd.DataFrame(inventory_records)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Inventory\Inventory.xlsx"
df.to_excel(output_path, index=False)
```
