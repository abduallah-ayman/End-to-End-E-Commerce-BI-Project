```python
import pandas as pd
import random
from datetime import timedelta
import numpy as np
```

```python
orders_df = pd.read_excel("D:\Portfolio\Data generation\Data\Order\Order.xlsx")

# Generate OrderID from row position
orders_df['OrderID'] = range(1, len(orders_df) + 1)
```

```python
status_rows = []

for _, row in orders_df.iterrows():
    order_id = row["OrderID"]
    order_date = row["OrderDate"]

    current_date = order_date

    status_flow = ["CREATED", "PAID", "SHIPPED", "DELIVERED"]

    for status in status_flow:
        if status == "CREATED":
            status_date = current_date
        elif status == "PAID":
            status_date = current_date + timedelta(days=random.randint(0, 2))
        elif status == "SHIPPED":
            status_date = current_date + timedelta(days=random.randint(1, 3))
        else:  # DELIVERED
            status_date = current_date + timedelta(days=random.randint(2, 5))

        status_rows.append({
            "OrderID": order_id,
            "Status": status,
            "StatusDate": status_date
        })

        current_date = status_date

df_status = pd.DataFrame(status_rows)
```

```python
df1, df2, df3, df4 = np.array_split(df_status, 4)

# Save each part to a separate Excel file
df1.to_excel("D:\Portfolio\Data generation\Data\Order\OrderStatusHistory_part1.xlsx", index=False)
df2.to_excel("D:\Portfolio\Data generation\Data\Order\OrderStatusHistory_part2.xlsx", index=False)
df3.to_excel("D:\Portfolio\Data generation\Data\Order\OrderStatusHistory_part3.xlsx", index=False)
df4.to_excel("D:\Portfolio\Data generation\Data\Order\OrderStatusHistory_part4.xlsx", index=False)
```
