## SubCategory

```python
subcategory_map = {
    1: ["Mobile Phones", "Laptops", "Accessories", "TVs"],
    2: ["Men Clothing", "Women Clothing", "Shoes", "Bags"],
    3: ["Furniture", "Kitchen Appliances", "Home Decor"],
    4: ["Fitness", "Camping", "Outdoor Gear"],
    5: ["Fiction", "Non-Fiction", "Educational"],
    6: ["Skincare", "Makeup", "Hair Care"],
    7: ["Board Games", "Puzzles", "Action Figures"],
    8: ["Car Accessories", "Motorbike Parts"],
    9: ["Medical Supplies", "Wellness"],
    10: ["Stationery", "Office Electronics"]
}

rows = []
```

```python
for category_id, subCategories in subcategory_map.items():
    for subCategory in subCategories:
        rows.append({
            "CategoryID" : category_id,
            "SubCategoryName" : subCategory
        })

df = pd.DataFrame(rows)
```

```python
output_path = "D:/Portfolio/Data generation/Data/ref_ProductSubCategory.xlsx"
df.to_excel(output_path, index=False)
```
