```python
import pandas as pd
import random
from datetime import timedelta
import numpy as np
```

```python
NUM_PAYMENTS = 800000  # PaymentIDs from 1 to 800,000
FAILED_PERCENT = 0.006  # 

# Define flows
completed_flow = ["INITIATED", "AUTHORIZED", "COMPLETED"]
failed_flow = ["INITIATED", "AUTHORIZED", "FAILED"]
```

```python
status_rows = []

for payment_id in range(1, NUM_PAYMENTS + 1):
    current_date = pd.Timestamp.now() - pd.Timedelta(days=random.randint(0, 730))  # random start date in last 2 years

    # Decide if this payment fails
    is_failed = random.random() < FAILED_PERCENT

    if is_failed:
        # First attempt fails
        for status in failed_flow:
            increment_days = random.randint(0, 3)
            current_date += pd.Timedelta(days=increment_days)
            status_rows.append({
                "PaymentID": payment_id,
                "Status": status,
                "StatusDate": current_date
            })

        # Second attempt succeeds (same PaymentID)
        retry_date = current_date + pd.Timedelta(days=1)
        for status in completed_flow:
            increment_days = random.randint(0, 3)
            retry_date += pd.Timedelta(days=increment_days)
            status_rows.append({
                "PaymentID": payment_id,
                "Status": status,
                "StatusDate": retry_date
            })
    else:
        # Normal completed payment
        for status in completed_flow:
            increment_days = random.randint(0, 3)
            current_date += pd.Timedelta(days=increment_days)
            status_rows.append({
                "PaymentID": payment_id,
                "Status": status,
                "StatusDate": current_date
            })

df_payment_status = pd.DataFrame(status_rows)
```
