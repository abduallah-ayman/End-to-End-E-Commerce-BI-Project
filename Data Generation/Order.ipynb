```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta
```

```python
fake = Faker()

NUM_ORDERS = 800000
NUM_CUSTOMERS = 50000
```

```python
orders = []

for _ in range(NUM_ORDERS):
    orders.append({
        "CustomerID": random.randint(1, NUM_CUSTOMERS),
        "OrderDate": fake.date_time_between(
            start_date="-2y",
            end_date="now"
        ),
        "TotalAmount": 0
    })

df = pd.DataFrame(orders)
```

```python
output_path = "D:\Portfolio\Data generation\Data\Order\Order.xlsx"
df.to_excel(output_path, index=False)
```
