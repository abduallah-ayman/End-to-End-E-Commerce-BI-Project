```python
ATTRIBUTE_RULES = {
    "Mobile Phones": {
        "Brand": ["Samsung", "Apple", "Xiaomi", "Oppo"],
        "Storage": ["64GB", "128GB", "256GB"],
        "RAM": ["4GB", "6GB", "8GB", "12GB"],
        "Color": ["Black", "White", "Blue", "Green"],
        "Screen Size": ["6.1\"", "6.5\"", "6.8\""]
    },
    "Laptops": {
        "Brand": ["Dell", "HP", "Lenovo", "Asus", "Apple"],
        "RAM": ["8GB", "16GB", "32GB"],
        "Storage": ["256GB SSD", "512GB SSD", "1TB SSD"],
        "Processor": ["i5", "i7", "Ryzen 5", "Ryzen 7"],
        "Screen Size": ["13\"", "15.6\"", "17\""]
    },
    "Men Clothing": {
        "Size": ["S", "M", "L", "XL"],
        "Color": ["Black", "White", "Blue", "Gray"],
        "Material": ["Cotton", "Polyester", "Denim"],
        "Fit": ["Regular", "Slim"]
    },
    "Shoes": {
        "Size": ["40", "41", "42", "43", "44"],
        "Color": ["Black", "White", "Red"],
        "Material": ["Leather", "Mesh", "Synthetic"],
        "Gender": ["Men", "Unisex"]
    },
    "TVs": {
        "Brand": ["Samsung", "LG", "Sony"],
        "Screen Size": ["43\"", "50\"", "55\"", "65\""],
        "Resolution": ["4K", "Full HD"],
        "Smart TV": ["Yes", "No"]
    },
    # default fallback
    "DEFAULT": {
        "Color": ["Black", "White", "Gray"],
        "Warranty": ["6 Months", "1 Year", "2 Years"]
    }
}
```

```python
import pandas as pd
import random

products = pd.read_excel("D:\Portfolio\Data generation\Data\Product\Product.xlsx")
products.reset_index(drop=True, inplace=True)
products["ProductID"] = products.index + 1
```

```python
subcategories = pd.read_excel("D:\Portfolio\Data generation\Data\Product\ProductSubCategory.xlsx")
subcategories.reset_index(drop=True, inplace=True)
subcategories["ID"] = subcategories.index + 1
```

```python
subcategories
```

```python
subcat_lookup = dict(
    zip(subcategories["ID"], subcategories["SubCategoryName"])
)
```

```python
attributes = []

for _, product in products.iterrows():
    product_id = product["ProductID"]
    subcat_id = product["SubCategoryID"]

    subcat_name = subcat_lookup[subcat_id]
    rules = ATTRIBUTE_RULES.get(subcat_name, ATTRIBUTE_RULES["DEFAULT"])

    min_attrs = min(3, len(rules))
    max_attrs = min(6, len(rules))

    attr_count = random.randint(min_attrs, max_attrs)
    chosen_attrs = random.sample(list(rules.keys()), attr_count)

    for attr in chosen_attrs:
        attributes.append({
            "ProductID": product_id,
            "AttributeName": attr,
            "AttributeValue": random.choice(rules[attr])
        })

df = pd.DataFrame(attributes)
```
