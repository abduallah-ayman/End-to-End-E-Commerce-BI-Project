```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()

NUM_CUSTOMERS = 50000
contacts = []
```

```python
for customer_id in range(1, NUM_CUSTOMERS + 1):
    # Mandatory primary email
    contacts.append({
        "CustomerID": customer_id,
        "Email": fake.unique.email(),
        "Phone": fake.phone_number(),
        "IsPrimary": 1
    })

    # Optional secondary contact (30% chance)
    if random.random() < 0.3:
        contacts.append({
            "CustomerID": customer_id,
            "Email": fake.unique.email(),
            "Phone": fake.phone_number(),
            "IsPrimary": 0
        })

df = pd.DataFrame(contacts)
```
