```python
import pandas as pd
import random
from datetime import timedelta
```

```python
# Load orders with OrderID and OrderDate
orders_df = pd.read_excel("D:\Portfolio\Data generation\Data\Order\Order.xlsx")
NUM_ORDERS = len(orders_df)

# Assume 5 payment methods
NUM_PAYMENT_METHODS = 5
```

```python
payments = []

for idx, row in orders_df.iterrows():
    order_id = idx + 1
    order_date = row["OrderDate"]
    total_amount = row["TotalAmount"]  # Currently 0, can update later

    # Decide if payment is full or partial (70% full, 30% partial)
    is_partial = random.choices([True, False], weights=[30, 70])[0]
    if is_partial:
        amount = round(random.uniform(1, max(total_amount, 1)), 2)
    else:
        amount = total_amount

    payment_method_id = random.randint(1, NUM_PAYMENT_METHODS)

    # PaymentDate after OrderDate
    payment_date = order_date + timedelta(days=random.randint(0, 5))

    payments.append({
        "OrderID": order_id,
        "PaymentMethodID": payment_method_id,
        "Amount": amount,
        "PaymentDate": payment_date
    })

df_payments = pd.DataFrame(payments)
```
